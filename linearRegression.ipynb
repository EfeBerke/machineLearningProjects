```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('seaborn-v0_8')

# x_train is the input(feature) variable (size in 1000 square feet)
# y_train is the target (price in 1000s of dollars)
x_train = np.array([1.0, 2.0])
y_train = np.array([300.0, 500.0])
print(f"x_train = {x_train}")
print(f"y_train = {y_train}")
```

```python
# m is the number of training examples
print(f"x_train.shape: {x_train.shape}")
m = x_train.shape[0] # In 1D array shape[0] gives the number of input variables
print(f"Number of training examples is: {m}")
```

```python
i = 0 
j = 1

# (x^(0), y^(0))
x_i = x_train[i]
y_i = y_train[i]

# (x^(1), y^(1))
x_j = x_train[j]
y_j = y_train[j]
print(f"(x^({i}), y^({i})) = ({x_i}, {y_i})")
print(f"(x^({j}), y^({j})) = ({x_j}, {y_j})")
```

```python
# Plot the data points
plt.scatter(x_train, y_train, marker='x', c='r') # c indicates color(r is red), marker indicates scatter's shape which is 'x'
# Set the title
plt.title("Housing Prices")
# Set the y-axis label
plt.ylabel('Price (in 1000s of dollars)')
# Set the x-axis label
plt.xlabel('Size (1000 sqft)')
plt.show()
```

```python
w = 200
b = 100
print(f"w: {w}")
print(f"b: {b}")
```

```python
def compute_model_output(x, w, b):
    """
    Computes the prediction of a linear model
    Args:
      x (ndarray (m,)): Data, m examples 
      w,b (scalar)    : model parameters  
    Returns
      f_wb (ndarray (m,)): model prediction
    """
    m = x.shape[0]
    f_wb = np.zeros(m) # Preallocating a numpy array with m elements
    for i in range(m):
        f_wb[i] = w * x[i] + b # Adding predicted target values to f_wb array
        
    return f_wb
```

```python
tmp_f_wb = compute_model_output(x_train, w, b)

# Plot our model prediction
plt.plot(x_train, tmp_f_wb, c='b',label='Our Prediction') # Plotting a line (x variables -> x_train, y_variables -> tmp_f_wb)

# Plot the data points
plt.scatter(x_train, y_train, marker='x', c='r',label='Actual Values')

# Set the title
plt.title("Housing Prices")
# Set the y-axis label
plt.ylabel('Price (in 1000s of dollars)')
# Set the x-axis label
plt.xlabel('Size (1000 sqft)')
plt.legend()
plt.show()
```
